# Automatic Object detection

```python
# Import system modules
import arcpy
from arcpy.ia import *
import warnings
warnings.filterwarnings('ignore')

# Check out the ArcGIS Image Analyst extension license
arcpy.CheckOutExtension("ImageAnalyst")
```

```python
PATH = r"G:\Final_Project\three_Labeled_Img"
```

```python
import os

categories = os.listdir(PATH)
categories
```

```python
os.listdir(os.path.join(PATH, 'advanced'))
```

```python
from tqdm import tqdm
```

```python
# Use the pre-trained model to each image and store the ouptuts

for i in range(3):
    for img in tqdm(os.listdir(os.path.join(PATH, categories[i]))):
        # object detection
        try:
            DetectObjectsUsingDeepLearning(r"G:\Final_Project\three_Labeled_Img\{}\{}".format(categories[i], img), 
                                           r"G:\Final_Project\pre_trained\MyProject\MyProject.gdb\{}".format(str(categories[i]) + img.strip('.png')), 
                                           r"G:\Final_Project\pre_trained\usa_building_footprints.dlpk", 
                                           "padding 128; threshold 0.82; batch_size 4; tile_size 512", 
                                           "NO_NMS", 
                                           "Confidence", 
                                           "Class", 
                                           0,
                                           "PROCESS_AS_MOSAICKED_IMAGE")
        except:
            pass
        
        # conversion
        arcpy.env.workspace = r"G:\Final_Project\pre_trained\MyProject\MyProject.gdb"
        arcpy.TableToTable_conversion("{}.dbf".format(str(categories[i]) + img.strip('.png')), 
                                      "G:\Final_Project\detected_output", 
                                      "{}.csv".format(str(categories[i]) + img.strip('.png')))
```
